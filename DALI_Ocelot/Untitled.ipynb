```python
import matplotlib.pyplot as plt

# Define figure and axis
fig, ax = plt.subplots(figsize=(10, 7))

# Draw nodes and connections manually
ax.plot([0.2, 0.5], [0.8, 0.8], 'k-', lw=2)  # EPICS Control -> Machine Settings
ax.plot([0.5, 0.8], [0.8, 0.5], 'k-', lw=2)  # Machine Settings -> Beam Diagnostics
ax.plot([0.8, 0.5], [0.5, 0.2], 'k-', lw=2)  # Beam Diagnostics -> ELEGANT Input
ax.plot([0.5, 0.2], [0.2, 0.5], 'k-', lw=2)  # ELEGANT Input -> ELEGANT Simulation
ax.plot([0.2, 0.5], [0.5, 0.8], 'k-', lw=2)  # ELEGANT Simulation -> Optimization Results
ax.plot([0.5, 0.5], [0.2, 0.8], 'k-', lw=2, ls='--')  # Optimization Results -> EPICS Control

# Add labels
ax.text(0.2, 0.85, "EPICS Control", fontsize=12, ha='center', bbox=dict(boxstyle="round", facecolor="lightblue", edgecolor="black"))
ax.text(0.5, 0.85, "Machine Settings\n(Magnets, RF, etc.)", fontsize=10, ha='center', bbox=dict(boxstyle="round", facecolor="lightblue", edgecolor="black"))
ax.text(0.85, 0.5, "Beam Diagnostics\n(Position, Current)", fontsize=10, ha='center', bbox=dict(boxstyle="round", facecolor="lightblue", edgecolor="black"))
ax.text(0.5, 0.15, "ELEGANT Input", fontsize=10, ha='center', bbox=dict(boxstyle="round", facecolor="lightblue", edgecolor="black"))
ax.text(0.15, 0.5, "ELEGANT Simulation", fontsize=10, ha='center', bbox=dict(boxstyle="round", facecolor="lightblue", edgecolor="black"))
ax.text(0.5, 0.5, "Optimization Results", fontsize=10, ha='center', bbox=dict(boxstyle="round", facecolor="lightblue", edgecolor="black"))

# Adjust axis
ax.axis('off')

# Save the image
plt.savefig("epics_elegant_workflow.png", dpi=300, bbox_inches='tight')
plt.show()
```
